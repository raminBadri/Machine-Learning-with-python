# Simple Linear Regression

## Import required packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score
```

## Read and look at the data from CSV file

```python
df = pd.read_csv("FuelConsumption.csv")
df.head()
```

```python
df.describe()
```

## Select three independent and the dependent features

```python
cdf = df[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
cdf.head(9)
```

```python
viz = cdf[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
viz.hist()
plt.show()
```

## Try to find the best feature to train the model

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("EMISSION")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("ENGINE SIZE")
plt.ylabel("EMISSION")
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("CYLINDERS")
plt.ylabel("EMISSION")
plt.show()
```

## Creating train and test samples from the selected data frame

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
print (len(train), len(test))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue', marker='s', label='Train')
ax1.scatter(test.ENGINESIZE, test.CO2EMISSIONS, color='red', marker='*', label='Test')
plt.xlabel("ENGINE SIZE")
plt.ylabel("EMISSION")
plt.legend(loc='upper left')
plt.show()
```

## Training the model with the selected feature

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[["ENGINESIZE"]])
train_y = np.asanyarray(train[["CO2EMISSIONS"]])
regr.fit(train_x, train_y)
theta_zero = regr.intercept_
theta_one = regr.coef_
print ("Coefficient: ", theta_one)
print ("Intecept: ", theta_zero)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.plot(train_x, theta_one[0][0]*train_x + theta_zero, 'r')
plt.xlabel("ENGINE SIZE")
plt.ylabel("EMISSION")
```

# Evaluation

```python
test_x = np.asanyarray(test[["ENGINESIZE"]])
test_y = np.asanyarray(test[["CO2EMISSIONS"]])
test_y_ = regr.predict(test_x)
print ("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print ("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print ("R2-score: %.2f" % r2_score(test_y, test_y_))
```
